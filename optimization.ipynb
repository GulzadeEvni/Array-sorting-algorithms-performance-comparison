```python
import random
import time
```

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1


def interpolation_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high and arr[low] <= target <= arr[high]:
        mid = low + ((target - arr[low]) * (high - low)) // (arr[high] - arr[low])
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

def jump_search(arr, target):
    n = len(arr)
    step = int(n**0.5)
    prev = 0

    while arr[min(step, n) - 1] < target:
        prev = step
        step += int(n**0.5)

        if prev >= n:
            return -1

    for i in range(prev, min(step, n)):
        if arr[i] == target:
            return i

    return -1
```

```python
def generate_sorted_array(size):
    unique_elements = set()
    sorted_array = []

    while len(sorted_array) < size:
        element = random.randint(0, size)
        if element not in unique_elements:
            unique_elements.add(element)
            sorted_array.append(element)

    sorted_array.sort()
    return sorted_array

def measure_search_time():
    sizes = [10000, 50000, 100000,300000, 500000, 1000000, 5000000, 10000000,15000000]

    for size in sizes:
        sorted_array = generate_sorted_array(size)

        print(f"Dizi Boyutu: {size}")
        print("-" * 40)

        # (first, middle, last, random)
        target_first = sorted_array[0]
        target_middle = sorted_array[size // 2]
        target_last = sorted_array[-1]
        target_random = random.choice(sorted_array)

        target_values = {
            "First Element": target_first,
            "Middle Element": target_middle,
            "Last Element": target_last,
            "Random Element": target_random
        }

        for target_name, target_value in target_values.items():
            results = []

            # zaman
            for algorithm in [linear_search, binary_search, interpolation_search, jump_search]:
                arr_copy = sorted_array[:]

                start_time = time.time()
                index = algorithm(arr_copy, target_value)
                end_time = time.time()
                execution_time = end_time - start_time

                results.append({
                    "Algorithm": algorithm.__name__,
                    "Index": index,
                    "Execution Time": execution_time
                })

            # zaman sıralama
            results.sort(key=lambda x: x["Execution Time"])

            # sonuç
            for result in results:
                algorithm_name = result["Algorithm"]
                index = result["Index"]
                execution_time = result["Execution Time"]

                print(f"Sıralı Dizi, {algorithm_name}, Hedef Eleman: {target_name}, Index: {index}, Süre: {execution_time:.6f} saniye")

            print()

measure_search_time()
```
